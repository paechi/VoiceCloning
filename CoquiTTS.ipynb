```python
from TTS.api import TTS

# Running a multi-speaker and multi-lingual model
TTS.list_models()
```

#### CoquiTTS provides a multilingual model called YourTTS, but for now it only supports English, French and Portuguese languages. Similar to TortoiseTTS, we can do the voice cloning for the English language.

```python
tts = TTS(model_name="tts_models/multilingual/multi-dataset/your_tts", progress_bar=False, gpu=True)
tts.tts_to_file("This is voice cloning.", speaker_wav="audios/6.wav", language="en", file_path="output.wav")
```

```python
import IPython.display as ipd
# print("original")
# ipd.display(ipd.Audio('audios/6.wav'))
print("cloned")
ipd.display(ipd.Audio('output.wav'))
```

#### Or, we can do voice cloning using one of the TTS models for English language and test it on our English audiofile(audios/6.wav)

```python
tts = TTS('tts_models/en/ljspeech/vits')
tts.tts_with_vc_to_file(
    "Unlike last week, the weather is really nice today!",
    speaker_wav="audios/6.wav",
    file_path="output.wav"
)
```

```python
import IPython.display as ipd
ipd.display(ipd.Audio('output.wav'))
```

####The generated speech sounds slightly robotic.

#### Now let's try to run the same reference audio on Ukrainian TTS model and check if it generate the the given Slavic text.

```python
tts = TTS('tts_models/uk/mai/vits') # Load vits model for Ukrainian language since no models for Russian language
tts.tts_with_vc_to_file(
    "Сегодня погода отличная, в отличии от прошлой недели",
    speaker_wav="audios/6.wav",
    file_path="output.wav"
)
```

```python
import IPython.display as ipd
ipd.display(ipd.Audio('output.wav'))
```

####The resulting speech has a strong English accent, but the voice is clear and speech is comprehensible.

#### Overall, CoquiTTS could be very useful for the multilingual voice cloning, but the code should be optimized, such that one model clones the voice and generates texts in all languages.
